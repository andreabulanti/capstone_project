```python
import pandas as pd
import numpy as np

world_bank = pd.read_excel("WorldBank.xlsx").rename({"Contry Name":"Country"}, axis=1)

world_bank["Population (M)"] = (world_bank["GDP (USD)"]/world_bank["GDP per capita (USD)"])/1000000

world_bank.head()
```

```python
world_bank.info()
```

```python
world_bank.describe()
```

```python
hdi = pd.read_csv("HDI.csv")

wb_hdi_2014 = (world_bank.query("Year == 2014")
               .merge(hdi[["iso3", "hdi_2014"]], how="left", left_on="Country Code", right_on = "iso3"))
              

wb_hdi_2014.head(20)
```

```python
wb_hdi_2014.describe()
```

# **_Preparing Data For Visualizazion_**

```python
gdp_pivot = world_bank.pivot_table(index="Year", columns="Region", values="GDP (USD)", aggfunc="sum")
pop_pivot = world_bank.pivot_table(index="Year", columns="Region", values="Population (M)", aggfunc="sum")
```

```python
gdp_pivot.head()
```

```python
pop_pivot.head()
```

```python
wb_hdi_by_region = wb_hdi_2014.groupby("Region")[["hdi_2014"]].mean().sort_values(by="hdi_2014", ascending = False)
```

```python
wb_hdi_by_region
```

# _Visualizazion_ 

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()
(ax.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending = False).index],
    labels = gdp_pivot.iloc[-1].sort_values(ascending = False).index)
)

ax.legend(loc="upper left")
ax.set_title("GDP has Grown Exponentially Over Time")
ax.set_ylabel("GDP (Trillions)")
```

```python
fig, ax = plt.subplots()
(ax.stackplot(
    pop_pivot.index,
    [pop_pivot[region] / 1_000 for region in pop_pivot.iloc[-1].sort_values(ascending = False).index], #vedi sopra
    labels = pop_pivot.iloc[-1].sort_values(ascending = False).index)
)

ax.legend(loc="upper left")
ax.set_title("Population has surged from 2 billions to 7.5 billion")
ax.set_ylabel("Population (Billions)")
```

```python
import matplotlib.ticker as mticker

minsize = min(wb_hdi_2014["Population (M)"])
maxsize = max(wb_hdi_2014["Population (M)"])

fig, ax = plt.subplots()

sns.scatterplot(
    data=wb_hdi_2014,
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    size="Population (M)",
    sizes=(minsize,maxsize),
    hue="Region",
    ax = ax,
).set(title="Life Expectancies increases as Countries get richer", yscale="log")

ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
```

```python
fig, axes = plt.subplots()

(sns.barplot(data = wb_hdi_by_region, x= wb_hdi_by_region.index, y="hdi_2014", ax=axes, label= wb_hdi_by_region.index)
 .set(title="HDI by Region", ylabel="HDI - Human Development Index")
)

axes.set_xticklabels(axes.get_xticklabels(), rotation=90)

plt.show;
```

```python
sns.scatterplot(data=wb_hdi_2014,
               x="Electric power consumption (kWh per capita)",
               y="GDP per capita (USD)",
                hue = "hdi_2014",
                palette="coolwarm_r"
               ).set(title="Electricity Drives Development")
```

```python
import matplotlib.gridspec as gridspec
import matplotlib as mpl

mpl.rcParams["font.size"] = 8


fig = plt.figure(figsize=(8,10))

gs = gridspec.GridSpec(nrows=12,ncols=6)


stack1_list = ["#66C2A5", "#fC8D62","#8DA0CB", "#E78AC3", "#A6D854", "#FFD92F", "#B3B3B3" ]
stack2_list = ["#66C2A5", "#A6D854", "#B3B3B3", "#fC8D62", "#E78AC3", "#8DA0CB", "#FFD92F"]
bar_list = ["#8DA0CB", "#fC8D62", "#E78AC3","#FFD92F","#66C2A5", "#A6D854", "#B3B3B3"]
bubble_list = ["#A6D854", "#fC8D62", "#FFD92F", "#66C2A5", "#B3B3B3", "#E78AC3", "#8DA0CB"]

fig.suptitle(
    "Report: Global Economic Report Development from 1960-2018",
    fontsize=14,
    fontweight="bold",
    x= .65,
    y= .93       
)


ax1 = fig.add_subplot(gs[0:4, 0:3])
ax1.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending = False).index],
    labels = gdp_pivot.iloc[-1].sort_values(ascending = False).index,
    colors=stack1_list)
ax1.set_title("GDP has Grown Exponentially Over Time", fontsize = 8)
ax1.set_ylabel("GDP (Trillions)")


ax2 = fig.add_subplot(gs[0:4, 3:6])
ax2.stackplot(
    pop_pivot.index,
    [pop_pivot[region] / 1_000 for region in pop_pivot.iloc[-1].sort_values(ascending = False).index],
    labels = pop_pivot.iloc[-1].sort_values(ascending = False).index,
    colors=stack2_list)
ax2.set_title("Population has surged from 2 billions to 7.5 billion",  fontsize = 8)
ax2.set_ylabel("Population (Billions)")
ax2.legend(bbox_to_anchor = (1.75,1), fontsize = 7)


ax3 = fig.add_subplot(gs[4:8, :])
sns.scatterplot(
    data=wb_hdi_2014,
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    size="Population (M)",
    sizes=(minsize,maxsize),
    hue="Region",
    palette= bubble_list,
    ax = ax3,
).set(title="Life Expectancies increases as Countries get richer", yscale="log")
ax3.yaxis.set_major_formatter(mticker.ScalarFormatter())
handles, labels = ax3.get_legend_handles_labels()
entries_to_skip = len(wb_hdi_2014["Region"].unique())+1

for h in handles[1:]:
    sizes= [s / 1.5 for s in h.get_sizes()]
    h.set_sizes(sizes)
    
ax3.legend(handles[entries_to_skip:], 
           labels[entries_to_skip:], 
           bbox_to_anchor=(0,.9), 
           loc=2, 
           borderaxespad=0.,
           frameon = False,
           ncols=6,
           fontsize=8
)


ax4 = fig.add_subplot(gs[8:12, 0:3])
(sns.barplot(data = wb_hdi_by_region, x= wb_hdi_by_region.index, y="hdi_2014", ax=ax4, palette=bar_list)
 .set(title="HDI by Region", ylabel="HDI - Human Development Index")
)


ax4.set_xticklabels(axes.get_xticklabels(), rotation=90)

ax5 = fig.add_subplot(gs[8:12, 3:6])
sns.scatterplot(data=wb_hdi_2014,
               x="Electric power consumption (kWh per capita)",
               y="GDP per capita (USD)",
                hue = "hdi_2014",
                palette="coolwarm_r",
                ax=ax5
               ).set(title="Electricity Drives Development")

fig.subplots_adjust(wspace=4, hspace=4)
```
